```python
import pandas as pd

# Load your stock market dataset into a DataFrame
# Ensure you have columns 'date', 'ticker', 'open', 'high', 'low', 'close', 'volume'
Stocks = pd.read_csv('30_min_candle_data_20200206_20230210.csv')

# Convert the 'date' column to datetime format
Stocks['date'] = pd.to_datetime(Stocks['date'])

# Now you can use the .dt accessor to extract the year
Stocks['year'] = Stocks['date'].dt.year

# Define the profit threshold (e.g., 5%)
profit_threshold = 5

# Sort the data by date
Stocks = Stocks.sort_values('date')

# Create a DataFrame to store predictions
predictions = pd.DataFrame()

# Group the data by 'ticker'
grouped = Stocks.groupby('ticker')

# Iterate through ticker groups
for ticker, group in grouped:
    annual_returns = []

    for year in [2020, 2021, 2022, 2023]:
        # Filter data for the specific year
        year_data = group[group['year'] == year]
        
        # Check if there is data available for the year and ticker
        if not year_data.empty:
            start_price = year_data['close'].iloc[0]
            end_price = year_data['close'].iloc[-1]
            annual_return = ((end_price / start_price) - 1) * 100
            is_profitable = annual_return > profit_threshold
        else:
            annual_return = 0  # Placeholder for missing data
            is_profitable = False  # Placeholder for missing data

        annual_returns.append({
            'Year': year,
            'Return': annual_return,
            'Is_Profitable': is_profitable
        })

    # Create a DataFrame for this stock's predictions
    stock_predictions = pd.DataFrame(annual_returns)
    stock_predictions['Ticker'] = ticker

    # Append the stock's predictions to the main predictions DataFrame
    predictions = pd.concat([predictions, stock_predictions])

# Select the top 4 profitable stocks for each year
top_stocks_2020 = predictions[predictions['Year'] == 2020].nlargest(4, 'Return')
top_stocks_2021 = predictions[predictions['Year'] == 2021].nlargest(4, 'Return')
top_stocks_2022 = predictions[predictions['Year'] == 2022].nlargest(4, 'Return')
top_stocks_2023 = predictions[predictions['Year'] == 2023].nlargest(4, 'Return')

print(top_stocks_2020)
print(top_stocks_2021)
print(top_stocks_2022)
print(top_stocks_2023)
```

```python
import pandas as pd

# Assuming your dataset is in a CSV file
Stocks = pd.read_csv('30_min_candle_data_20200206_20230210.csv')

# Convert the 'date' column to a pandas datetime object
Stocks['date'] = pd.to_datetime(Stocks['date'])

# Set the 'date' column as the index
Stocks.set_index('date', inplace=True)

# Calculate the daily returns for each ticker
Stocks['daily_return'] = Stocks.groupby('ticker')['close'].pct_change()

# Resample the data to get monthly returns
Stocks_monthly = Stocks.groupby(['ticker', pd.Grouper(freq='M')])['daily_return'].agg(lambda x: (1 + x).prod() - 1).reset_index()

# Correct the column name
Stocks_monthly.rename(columns={'daily_return': 'monthly_return'}, inplace=True)

# Pivot the data to have tickers as columns and months as rows
Stocks_monthly_pivot = Stocks_monthly.pivot(index='date', columns='ticker', values='monthly_return')
print(Stocks_monthly)
Stocks_monthly.head(5)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with features and target variable
# Features
features = Stocks[['open', 'high', 'low', 'volume']]

# Target Variable
target = Stocks['close']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Normalize the data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape data for LSTM (assuming a time series structure)
X_train_reshaped = np.reshape(X_train_scaled, (X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_reshaped = np.reshape(X_test_scaled, (X_test_scaled.shape[0], X_test_scaled.shape[1], 1))

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train_reshaped.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train_reshaped, y_train, epochs=50, batch_size=32)

# Make predictions on the test set
y_pred = model.predict(X_test_reshaped)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')

# Visualize predictions vs. actual values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted Values')
plt.show()
```
